1. **Displaying the Items from Inventory**



```python
import json
import pandas as pd
fd = open("record.json",'r')
r = fd.read()
fd.close()

record = json.loads(r)
df = pd.DataFrame(record).T
df
```

2. **INCREASING quantity of EXISTING product**

```python
prod_id = str(input("Enter product id:"))
name = str(input("Enter name:"))
price = int(input("Enter price:"))
quantity = int(input("Enter quantity:"))

quantity=record[prod_id]['quantity']+quantity

record[prod_id] = {'name': name, 'price': price, 'quantity': quantity}

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```

4. **Add New Item into Inventory**

```python
prod_id = str(input("Enter product id:"))
name = str(input("Enter name:"))
price = int(input("Enter price:"))
quantity = int(input("Enter quantity:"))
if quantity<=0 or price<=0:
  print("invalid input")
else:
  record[prod_id] = {'name': name, 'price': price, 'quantity': quantity}

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```

5. **DELETE ELEMENT FROM RECORD**

```python
prod_id = str(input("Enter product id:"))
if prod_id in record:
  del record[prod_id]
  print(prod_id + "  prod_id is deleted")
else:
  print("prod_id is not available")  

```
